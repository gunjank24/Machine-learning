```python
import pandas as pd
data = pd.read_csv("cwa.csv")
data.head
```

```python
selected_data = data.iloc[:,1:3]
selected_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

k = 3
kmeans = KMeans(k)
kmeans.fit(selected_data)
```

```python
clusters = kmeans.fit_predict(selected_data)
clusters
```

```python
cluster_data = data.copy()
cluster_data['Clusters'] = clusters
plt.scatter(cluster_data['Income Classification'],  c=cluster_data['Clusters'],cmap='rainbow')
```
